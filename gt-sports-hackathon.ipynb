First load the libraries

```python
#load libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import data,filters
from skimage.measure import compare_ssim as ssim
```

Then import the video as image frames

```python
vs = cv2.VideoCapture("trainingdata.mp4")
frames = []
while True:
    (grabbed,tmp) = vs.read()
    frames.append(tmp)
    if not grabbed:
        break
```

Look at the similarity index between frames to find when the camera angle switches

```python
#get the similarity index for each frame
simIdx = np.zeros(np.shape(frames))
for i in np.arange(np.shape(frames)[0]):
    simIdx[i] = ssim(frames[i+1], frames[i], multichannel=True)
    
```

```python
#plot the results    
plt.plot(simIdx)
```

Check that the similarity index is capturing when the camera angle is changing

```python
switchFrames = [i for i, val in enumerate(np.diff(simIdx)>0.1) if val] 
switchFrames
```

```python
#get indices where the similarity diff is greater than 50%
switchFrames = [i for i, val in enumerate(np.diff(simIdx)>0.25) if val] 

fig, ax = plt.subplots(np.shape(switchFrames)[0],5, figsize=(20, 10))
for i in np.arange(np.shape(switchFrames)[0]):
    ax[i,0].imshow(frames[switchFrames[i]-2])
    ax[i,1].imshow(frames[switchFrames[i]-1])
    ax[i,2].imshow(frames[switchFrames[i]])
    ax[i,2].set_title('Switch frame')
    ax[i,3].imshow(frames[switchFrames[i]+1])
    ax[i,4].imshow(frames[switchFrames[i]+2])
    
    
cv2.cvtColor(frames[switchFrames[i]-2], cv2.COLOR_BGR2RGB)
```
